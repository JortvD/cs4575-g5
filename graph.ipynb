```python
import numpy
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the CSV file
data = pd.read_csv("output/2025-02-19_21-57-58.csv")
```

```python
data2 = pd.read_csv("output/2025-02-19_21-36-23.csv")
```

```python
# Plot the "PACKAGE_ENERGY (J)" column
plt.plot(data["PACKAGE_ENERGY (J)"])
plt.plot(data2["PACKAGE_ENERGY (J)"])
plt.xlabel("Index")
plt.ylabel("PACKAGE_ENERGY (J)")
plt.title("PACKAGE_ENERGY (J) over Time")
plt.show()
```

```python
# Calculate the gradient of PACKAGE_ENERGY (J)
package_energy_gradient = numpy.gradient(data["PACKAGE_ENERGY (J)"])
package_energy_gradient2 = numpy.gradient(data2["PACKAGE_ENERGY (J)"])

# Remove negative gradients
package_energy_gradient = package_energy_gradient[package_energy_gradient >= 0]
package_energy_gradient2 = package_energy_gradient2[package_energy_gradient2 >= 0]

sum = numpy.sum(package_energy_gradient)
sum2 = numpy.sum(package_energy_gradient2)
print(f"Sum of positive gradients: {sum/3600:.2f} Wh, {sum2/3600:.2f} Wh")
sum_time = numpy.sum(data["Delta"])
sum_time2 = numpy.sum(data2["Delta"])
print(f"Sum of time: {sum_time/1000:.2f}s, {sum_time2/1000:.2f}s")
print(f"Average power: {sum/(sum_time/1000):.2f} W, {sum2/(sum_time2/1000):.2f} W")

plt.plot(package_energy_gradient, label='Gradient 1')
plt.plot(package_energy_gradient2, label='Gradient 2')
plt.legend()
plt.xlabel("Index")
plt.ylabel("Gradient of PACKAGE_ENERGY (J)")
plt.title("Gradient of PACKAGE_ENERGY (J) over Time")
plt.show()
```

```python
# Filter columns that match the pattern "CPU_FREQUENCY_X"
cpu_frequency_columns = [col for col in data.columns if col.startswith("CPU_USAGE_")]

# Plot the filtered columns
data[cpu_frequency_columns].plot()
plt.xlabel("Index")
plt.ylabel("CPU Frequency")
plt.title("CPU Frequencies over Time")
plt.legend(title="CPU Cores")
plt.show()
```
